# solution 1
+ GET Method:

+ Purpose: The GET method is used to request data from a specified resource.
+ Data in URL: Parameters are appended to the URL.
+ Safe: Does not modify resources.
+ Typical Use: Fetching data (e.g., HTML pages, images).

+ POST Method:

+ Purpose: The POST method is used to send data to a server to create/update a resource.
+ Data in Body: Parameters are included in the body of the request.
+ Modifies Resources: Used to create/update data.
+ Typical Use: Submitting forms, uploading files.

# solution 2 
+ In Flask, the request object is used to access incoming request data. This includes form data, query parameters, headers, cookies, and file uploads. It's an essential part of handling HTTP requests in a web application.

# solution 3 
+ The redirect() function in Flask is used to redirect the client to a different URL. This is often used after a form submission to prevent resubmission on page refresh, or to navigate users to a different page after completing an action.

# solution 4
+ Templates:

+ Templates in Flask are HTML files that are used to generate dynamic web pages. They allow embedding of Python code within HTML using templating languages like Jinja2.
+ render_template() Function:

+ The render_template() function is used to render these templates, passing data from the backend to the front end. This function takes the template file and any data to be rendered within the template.
